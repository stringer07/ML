```python
import torch
import networkx as nx 
import matplotlib.pyplot as plt 
```

```python
#画图的函数
def visualize_graph(G,color):
    plt.figure(figsize=(7,7))
    plt.xticks([])
    plt.yticks([])
    nx.draw_networkx(G,pos=nx.spring_layout(G,seed=42),with_labels=False,node_color=color,cmap="Set2")
    plt.show()
#画点的函数
def visualize_embedding(h,color,epoch=None,loss=None):
    plt.figure(figsize=(7,7))
    plt.xticks([])
    plt.yticks([])
    h=h.detach().cpu().numpy()
    plt.scatter(h[:,0],h[:,1],s=140,c=color,cmap="Set2")
    if epoch is not None and loss is not None:
        plt.xlabel(f'Epoch: {epoch},Loss: {loss.item():.4f}',fontsize=16)
    plt.show()
```

```python
#空手道俱乐部，每个图34个点（只有一个图），每个点特征是一个34维向量 
#对每一个点进行分类任务 
from torch_geometric.datasets import KarateClub
dataset=KarateClub()
print(f'Dataset: {dataset}:')
print('================')
print(f'Number of graphs: {len(dataset)}')
print(f'Number of fetures: {dataset.num_features}')
print(f'Number of classes: {dataset.num_classes}')
```

```python
data=dataset[0]#只有一个图，提取第一个元素
print(data)
#空手道俱乐部，每个图34个点（只有一个图），每个点特征是一个34维向量
#edge_index 一定是2*N（N=34*34）
#y是标签
#train_mask 指定哪些点有标签，哪些点没有标签
```

```python
edge_index=data.edge_index
print(edge_index.t())
```

```python
from torch_geometric.utils import to_networkx
G=to_networkx(data,to_undirected=True)
visualize_graph(G,color=data.y)
```

```python
from torch.nn import Linear
from torch_geometric.nn import GCNConv

class GCN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        torch.manual_seed(1234)
        #只需定义好输入特征和输出特征即可，
        self.conv1=GCNConv(dataset.num_features,4)#输入的dataset.num_fetures是一个34维向量
        self.conv2=GCNConv(4,4)
        self.conv3=GCNConv(4,2)
        self.classifier=Linear(2,dataset.num_classes) #dataset.num_classes=4
    
    def forward(self,x,edge_index):#edge_index是邻接矩阵
        h=self.conv1(x,edge_index)#输入特征与邻接矩阵
        h=h.tanh()
        h=self.conv2(h,edge_index)
        h=h.tanh()
        h=self.conv3(h,edge_index)
        h=h.tanh()

        #分类层
        out=self.classifier(h)  #fc全连接层

        return out,h

model=GCN()
print(model)
```

```python
model=GCN()
_,h=model(data.x,data.edge_index)
print(f'Embedding shape: {list(h.shape)}')
visualize_embedding(h,color=data.y)
```

### 训练模型（semi-supervised）

```python
model=GCN()
criterion=torch.nn.CrossEntropyLoss() #定义损失函数
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)

def train(data):
    optimizer.zero_grad()
    out,h=model(data.x,data.edge_index)#h是两位向量 主要是画图
    loss=criterion(out[data.train_mask],data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return loss,h

for epoch in range(401):
    loss,h=train(data)
    if epoch%10==0:
        visualize_embedding(h,color=data.y,epoch=epoch,loss=loss)
```
